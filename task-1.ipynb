```python
import sklearn
print(sklearn.__version__)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('/content/task1 data.csv')
data.head()
```

```python
x=data.iloc[:,:-1].values
y=data.iloc[:,-1].values
print(x)
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.4,random_state=0)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train,y_train)
```

```python
y_predict=regressor.predict(x_test)
print(y_predict.shape)
```

```python
plt.scatter(x_test[:,0],y_test,color='red')
plt.scatter(x_test[:,0],regressor.predict(x_test))
plt.title("square area vs  price")
plt.xlabel("sqaure area")
plt.ylabel("price of house")
plt.show()
```

```python
plt.scatter(x_test[:,1],y_test,color='red')
plt.scatter(x_test[:,1],regressor.predict(x_test))
plt.title("Bathroom vs  price")
plt.xlabel("Bathroom")
plt.ylabel("price of house")
plt.show()
```

```python
plt.scatter(x_test[:,1],y_test,color='red')
plt.scatter(x_test[:,1],regressor.predict(x_test))
plt.title("Bedroom vs  price")
plt.xlabel("Bedroom")
plt.ylabel("price of house")
plt.show()
```
